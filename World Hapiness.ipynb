# Removing Outliers

```python
from scipy.stats import zscore
import numpy as np
z=np.abs(zscore(hapiness))
threshold=3
new_hapiness=hapiness[(z<3).all(axis=1)]
print(hapiness.shape)
print(new_hapiness.shape)
```

```python
new_hapiness.plot(kind='box',subplots=True,layout=(3,7),color='green',figsize=(13,6))
plt.tight_layout()
```

Here, "standard Error" has less correlation with the target column and there are a lot of outliers as well. So, Let delete this column from the dataset.

```python
new_hapiness.drop(['Standard Error'], axis = 1,inplace=True)
```

```python
new_hapiness.head()
```

# Remove skewness

```python
#Prepare the dataset for training the model
x=new_hapiness.drop('Happiness Rank',axis=1)
y=new_hapiness['Happiness Rank']
```

# yeo-johnson method

```python
from sklearn.preprocessing import power_transform     #box-cox is only for positive but yeo-johnson is for both negative and positive
x=power_transform(x,method='yeo-johnson')
```

```python
new_hapiness.columns
```

```python
import itertools
cols=['Country', 'Region', 'Happiness Rank', 'Happiness Score',
       'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)',
       'Freedom', 'Trust (Government Corruption)', 'Generosity',
       'Dystopia Residual']
length=len(cols)
cs=["b","r","g","c","m","k","lime"]
fig=plt.figure(figsize=(13,25))

for i,j,k in itertools.zip_longest(cols,range(length),cs):
    plt.subplot(3,4,j+1)
    ax=sns.distplot(new_hapiness[i],color=k,rug=True)
    ax.set_facecolor("w")
    plt.axvline(new_hapiness[i].mean(),linestyle="dashed",label="mean",color="k")
    plt.legend(loc="best")
    plt.title(i,color="navy")
    plt.xlabel("")
```

# Model Creating

# Best random state

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_t=sc.fit_transform(x)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
lss=Lasso()
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
for i in range(0,100):
    x_train,x_test,y_train,y_test=train_test_split(x_t,y,test_size=.42,random_state=i)
    lss.fit(x_train,y_train)
    pred_train=lss.predict(x_train)
    pred_test=lss.predict(x_test)
    if round(r2_score(y_train,pred_train)*100,2)==round(r2_score(y_test,pred_test)*100,2):
        print("At random state",i,"The model performs very well")
        print("At random state :-",i)
        print("Training r2_score is :-",r2_score(y_train,pred_train)*100)
        print("Testing r2_score is :-",r2_score(y_test,pred_test)*100)
```

```python
print("At random state",i,"The model performs very well")
print("At random state :-",i)
print("Training r2_score is :-",r2_score(y_train,pred_train)*100)
print("Testing r2_score is :-",r2_score(y_test,pred_test)*100)
```

# Cross Valodation score

```python
from sklearn.linear_model import Lasso

parameters={'alpha':[.0001,.001,.01,.1,1,10],'random_state':list(range(0,10))}
clf=GridSearchCV(lss,parameters)
clf.fit(x_train,y_train)
print(clf.best_params_)
```

```python
ls=Lasso(alpha=1,random_state=0)
ls.fit(x_train,y_train)
ls.score(x_train,y_train)
pred_ls=ls.predict(x_test)

lss=r2_score(y_test,pred_ls)
for j in range (2,30):
    lsscore=cross_val_score(ls,x_t,y,cv=j)
    lsc=lsscore.mean()
    print("At cv:-",j)
    print("Cross Validation Score is :- ", lsc*100)
    print("R2 score is :-", lss*100)
    print("\n")
```

At CV=4 we have got better cross validation score. So, I am taking CV=4

```python
print("Error")
print("Mean Absolute Eror : ",mean_absolute_error(y_test,pred_ls))
print("Mean Squared Eror : ",mean_squared_error(y_test,pred_ls))
print("Root Mean Squared Eror : ",np.sqrt(mean_squared_error(y_test,pred_ls)))
```
